# Plot data from sample file. 

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# Load Dummy Data (CSV File)
# This file's index is the Row number from MSexcel
data_read = pd.read_csv(r'C:\Users\drgdm\OneDrive - University of Puerto Rico\Universidad\Clases\Capstone Proyect\Project\Application\Random_Dummy_Data_v1.csv')
# data_read
```

```python
# Rearrange Index
data_read.set_index('INDEX').tail()
```

```python
data_read['S1'].tail(10000).plot(figsize=(25, 6))
plt.show()
```

```python
fig, ax = plt.subplots()
data_read.tail(1000).plot(x='INDEX', y='S1', ax=ax, legend=False, figsize=(25, 6))

plt.show()
```

# Multiple Lines in a Graph

```python
fig, ax = plt.subplots()
data_read.tail(100).plot(x='INDEX', y='S1', ax=ax, legend=False, figsize=(25, 6), label='Sensor 1')
data_read.tail(100).plot(x='INDEX', y='S2', ax=ax, legend=False, figsize=(25, 6), label='Sensor 2')
data_read.tail(100).plot(x='INDEX', y='S3', ax=ax, legend=False, figsize=(25, 6), label='Sensor 3')
data_read.tail(100).plot(x='INDEX', y='S4', ax=ax, legend=False, figsize=(25, 6), label='Sensor 4')

plt.title('Sensor Plot')
plt.legend()
plt.show()
```

```python
data_read.tail(100).plot(x='INDEX', y='S1', legend=False, figsize=(25, 6), label='Sensor 1')
data_read.tail(100).plot(x='INDEX', y='S2', legend=False, figsize=(25, 6), label='Sensor 2')
data_read.tail(100).plot(x='INDEX', y='S3', legend=False, figsize=(25, 6), label='Sensor 3')
data_read.tail(100).plot(x='INDEX', y='S4', legend=False, figsize=(25, 6), label='Sensor 4')

plt.title('Sensor Plot')
plt.legend()
plt.show()
```

# Multiple Lines | Plot in Other Window with Faster Rendering | Interactive

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import matplotlib

# Load Dummy Data (CSV File)
# This file's index is the Row number from MSexcel
data_read = pd.read_csv(r'C:\Users\drgdm\OneDrive - University of Puerto Rico\Universidad\Clases\Capstone Proyect\Project\Application\Random_Dummy_Data_v1.csv')
# Rearrange Index
data_read.set_index('INDEX').tail()

# Open in New Qt5 Interactive Window
matplotlib.use('Qt5Agg')
%matplotlib qt5

# Multiple Lines in same Plot.
fig, ax = plt.subplots()
data_read.plot(x='INDEX', y='S1', ax=ax, legend=False, figsize=(25, 6), label='Sensor 1')
data_read.plot(x='INDEX', y='S2', ax=ax, legend=False, figsize=(25, 6), label='Sensor 2')
data_read.plot(x='INDEX', y='S3', ax=ax, legend=False, figsize=(25, 6), label='Sensor 3')
data_read.plot(x='INDEX', y='S4', ax=ax, legend=False, figsize=(25, 6), label='Sensor 4')

#Faster Rendering
matplotlib.rcParams['path.simplify'] = True
matplotlib.rcParams['path.simplify_threshold'] = 1.0

plt.title('Sensor Plot')
plt.legend()
plt.show()
```
